```python
from langgraph.graph import StateGraph,START,END
from langchain_groq import ChatGroq
```

```python
# Import the os module to interact with the operating system
import os

# Import the userdata module from google.colab to access secrets
from google.colab import userdata

# Retrieve the Groq API key from Colab's Secrets Manager
# and set it as an environment variable
os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')
```

```python
model = ChatGroq(model="llama-3.3-70b-versatile", api_key=os.environ['GROQ_API_KEY'])
```

```python
from typing import TypedDict, Optional
class LLMState(TypedDict):
  question: str
  answer: Optional[str]
```

```python
def llm_qa(state: LLMState)->LLMState:
  question=state["question"]
  prompt  = f"Answer the following question: {question}"
  answer = model.invoke(prompt).content
  return {'answer': answer}
```

```python
graph =StateGraph(LLMState)
graph.add_node("llm_qa",llm_qa)
graph.add_edge(START,"llm_qa")
graph.add_edge("llm_qa",END)
app= graph.compile()
```

```python
initial_state={"question":"Who are you?"}
final_state=app.invoke(initial_state)
```
